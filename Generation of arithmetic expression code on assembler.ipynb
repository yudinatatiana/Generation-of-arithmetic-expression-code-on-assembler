1) Идём слева-направо
2) Используемые команды
mov src, dest - пересылка значения из dest в src
xchg dest, src = меняет местами значения в dest и src
push dest - помещение значения из dest в стек
pop src - выталкивание значения из стека в src
mul (add, sub) src, dest - умножение (сложение, вычитание) содержимого src на (с, из) dest. Результат помещается в src

```python
class my_stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def is_empty(self):
        return (self.items == [])
    
    def top(self):
        a = self.items.pop()
        self.items.append(a)
        return a
    
    def size(self):
        return len(self.items)
```

```python
import re
def alpha(a):
    if re.match('[a-zA-Z]', a) is not None:
        return True
    else:
        return False
def digit(a):
    if re.match('[0-9]', a) is not None:
        return True
    else:
        return False
```

```python
def func(s):
    i = 0
    s_out = ''
    stack = my_stack()
    s = s + ' '
    while i < len(s):    
        if alpha(s[i]):
            curr_temp = ''
            while alpha(s[i]) or digit(s[i]):
                curr_temp += s[i]
                s_out += s[i]
                i += 1
            s_out += ' '
            print('mov eax, ', curr_temp)
            print('push eax')
        elif digit(s[i]):
            curr_temp = ''
            while digit(s[i]):
                curr_temp += s[i]
                s_out += s[i]
                i += 1
            s_out += ' '
            print('mov eax, ', curr_temp)
            print('push eax')  
        elif s[i] == '+' or s[i] == '-':
            if stack.is_empty():
                stack.push(s[i])
                i += 1
            elif stack.top() == '+' or stack.top() == '-' or stack.top() == '*' or stack.top() == '/':
                print('pop eax', 'pop edx', sep='\n')
                if stack.top() == '+':
                    print('add eax, edx', 'push eax', sep='\n')
                elif stack.top() == '*':
                    print('imul eax, edx', 'push eax', sep='\n')
                elif stack.top() == '-':
                    print('xchg eax, edx', 'sub eax, edx', 'push eax',  sep='\n')
                elif stack.top() == '/':
                    print('xchg eax, edx', 'idiv eax, edx', 'push eax', sep='\n')
                s_out += stack.top()
                s_out += ' '
                stack.pop()
            elif stack.top() == '(':
                stack.push(s[i])
                i += 1          
        elif s[i] == '*' or s[i] == '/':
            if stack.is_empty():
                stack.push(s[i])
                i += 1
            elif stack.top() == '+' or stack.top() == '-':
                stack.push(s[i])
                i += 1
            elif stack.top() == '*' or stack.top() == '/':
                print('pop eax', 'pop edx', sep='\n')
                s_out += stack.top()
                s_out += ' '
                if stack.top() == '*':
                    print('imul eax, edx', 'push eax', sep='\n')
                elif stack.top() == '/':
                    print('xchg eax, edx', 'idiv eax, edx', 'push eax', sep='\n')
                stack.pop()
            elif stack.top() == '(':
                stack.push(s[i])
                i += 1
        elif s[i] == '(':
            stack.push(s[i])
            i += 1
        elif s[i] == ')':
            if stack.is_empty():
                print('Error')
                i = stack.size() + 1
            elif stack.top() == '(':
                i += 1
                stack.pop()
            elif stack.top() == '+' or stack.top() == '-' or stack.top() == '*' or stack.top() == '/':
                print('pop eax', 'pop edx', sep='\n')
                if stack.top() == '+':
                    print('add eax, edx', 'push eax', sep='\n')
                if stack.top() == '*':
                    print('imul eax, edx', 'push eax', sep='\n')
                if stack.top() == '-':
                    print('xchg eax, edx', 'sub eax, edx', 'push eax', sep='\n')
                if stack.top() == '/':
                    print('xchg eax, edx', 'idiv eax, edx', 'push eax', sep='\n')
                s_out += stack.top()
                s_out += ' '
                stack.pop()
        elif i+1 == len(s):
            if stack.is_empty():
                i += 1
            elif stack.top() == '+' or stack.top() == '-' or stack.top() == '*' or stack.top() == '/':
                print('pop eax', 'pop edx', sep='\n')
                if stack.top() == '+':
                    print('add eax, edx', 'push eax', sep='\n')
                if stack.top() == '*':
                    print('imul eax, edx', 'push eax', sep='\n')
                if stack.top() == '-':
                    print('xchg eax, edx', 'sub eax, edx', 'push eax', sep='\n')
                if stack.top() == '/':
                    print('xchg eax, edx', 'idiv eax, edx', 'push eax', sep='\n')
                s_out += stack.top()
                s_out += ' '
                stack.pop()
    print(s_out)
```
